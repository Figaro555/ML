```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('train.csv', index_col = 'id')
test_data = pd.read_csv('test.csv', index_col = 'id')
```

## Микита: перевіряємо на відсутність значень( в нашому змаганні це -1)

```python
num_features_null = train_data[train_data == -1].sum()
print(num_features_null)
```

```python
num_features_test_null = test_data[test_data == -1].sum()
print(num_features_test_null)
```

## Перевіряємо кореляційну матрицю на дуже велику кореляцію

```python
corrMatt = train_data.corr()
mask = np.array(corrMatt)
mask[np.tril_indices_from(mask)] = False
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMatt, cmap="Greens", mask=mask,vmax=.8, square=True)
```

```python
corrMatt[corrMatt > 0.7].count().sum() - corrMatt[corrMatt == 1].count().sum()
```

```python
for col_name in corrMatt.columns:
    for row_name in corrMatt.columns:
        if row_name in corrMatt and col_name in corrMatt:
            if  corrMatt[col_name][row_name]> 0.7 and corrMatt[col_name][row_name] < 1:            
                corrMatt = corrMatt.drop(row_name, 1)
                corrMatt = corrMatt.drop(row_name, 0)
                train_data = train_data.drop(row_name,1)
                test_data = test_data.drop(row_name,1)
            
                print("we delete " + row_name)
            
```

## Микола: Подивимося на розподіл кожної фічі та видалимо ті в яких багато значень в одній колонці (в тому числі нульові)

```python
train_data[train_data.select_dtypes(include=['float64', 'int64']).columns].head(1000).hist(figsize=(20, 40), color = 'g', bins=30, xlabelsize=8, ylabelsize=8)
```

```python
features_to_delete = ['ps_ind_05_cat','ps_ind_08_bin','ps_ind_09_bin','ps_ind_10_bin','ps_ind_11_bin','ps_ind_12_bin','ps_ind_13_bin','ps_ind_17_bin','ps_ind_18_bin','ps_car_02_cat','ps_car_04_cat','ps_car_07_cat','ps_car_10_cat','ps_calc_15_bin','ps_calc_20_bin']
for feature in features_to_delete:
    train_data = train_data.drop(feature,1)
    test_data = test_data.drop(feature,1)
```

```python
train_data[train_data.select_dtypes(include=['float64', 'int64']).columns].head(1000).hist(figsize=(20, 40), color = 'g', bins=30, xlabelsize=8, ylabelsize=8)
```

## Заповнимо нульові значення за допомогою forward fill, backward fill які лише повторюють значення попереднього рядка проте зважаючи що дані більш-менш випадково розкидані то від цього розподіли практично не змінюються

```python
num_features_test_null = train_data[train_data==-1].sum()
print(num_features_test_null)
```

```python
num_features_test_null = test_data[test_data == -1].sum()
print(num_features_test_null)
```

```python
for col_name in test_data.columns:
    train_data.loc[train_data[col_name] == -1, col_name] = np.nan
    test_data.loc[test_data[col_name] == -1, col_name] = np.nan
```

```python
tr_data = train_data.ffill().bfill()
te_data = test_data.ffill().bfill()
```

```python
num_features_null = tr_data.isnull().sum()
print(num_features_null)
```

```python
tr_data[tr_data.select_dtypes(include=['float64', 'int64']).columns].head(1000).hist(figsize=(20, 40), color = 'g', bins=30, xlabelsize=8, ylabelsize=8)
```

```python
train_data[train_data.select_dtypes(include=['float64', 'int64']).columns].head(1000).hist(figsize=(20, 40), color = 'g', bins=30, xlabelsize=8, ylabelsize=8)
```

## Микита: У змаганні позначка cat означає категоріальну змінну, оскільки ми не можемо їх порівняти то для кожного поля категоріальної змінної виділяємо окремий рядок

```python
for col_name in train_data.columns:
    if 'cat' in col_name:
        tr_data[col_name] = tr_data[col_name].astype("category")
        te_data[col_name] = te_data[col_name].astype("category")
        tr_data = pd.get_dummies(tr_data, columns = (tr_data.describe(include = "category")).columns , prefix =(tr_data.describe(include = "category")).columns  )
        te_data = pd.get_dummies(te_data, columns = (te_data.describe(include = "category")).columns , prefix =(te_data.describe(include = "category")).columns  )
```

# Почнемо тренування моделі

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn import metrics
```

## Микола: Спробуємо працювати з незбалансованими данними

```python
train_x, test_x, train_y, test_y = train_test_split(tr_data.drop('target',1),tr_data['target'] , test_size=0.05)
```

```python
sc = StandardScaler()
scaled_train_x = sc.fit_transform(train_x)
scaled_test_x = sc.transform(test_x)
```

```python
mlp = MLPClassifier(random_state=1, max_iter=100)
mlp.fit(scaled_train_x, train_y)
```

```python
roc_auc_score(test_y,mlp.predict_proba(scaled_test_x)[:,1])
```

```python
metrics.plot_roc_curve(mlp, test_x, test_y)
plt.show()                      
```

```python
result = mlp.predict_proba(sc.transform(te_data))[:, 1]
res = te_data.copy()
res = res.assign(target = result)
for f in res.columns:
    if f!= 'target':
        res = res.drop(f,1)
```

```python
res.to_csv('res.csv', sep=',')
```

## Спробуємо логістичну регресію зі кроссвалідацією

```python
lr = LogisticRegressionCV(cv = 5, max_iter = 300)
lr.fit(scaled_train_x, train_y)
```

```python
roc_auc_score(test_y,lr.predict_proba(scaled_test_x)[:,1])
```

```python
metrics.plot_roc_curve(mlp, scaled_test_x, test_y)
plt.show()   
```

```python
result1 = lr.predict_proba(sc.transform(te_data))[:, 1]
res = te_data.copy()
res = res.assign(target = result1)
for f in res.columns:
    if f!= 'target':
        res = res.drop(f,1)
```

```python
res.to_csv('res1.csv', sep=',')
```
